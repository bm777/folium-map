```python
url = "https://kandaweather-mainnet.ddns.net/v1/chain/get_table_rows"
HEADER = {
            "content-type": "application/json", "Accept-Charset": "UTF-8"
        }
DATA = "{\"code\":\"dclimateiot4\",\"table\":\"weather\",\"scope\":\"dclimateiot4\",\"index_position\":\"first\",\"json\":\"true\",\"limit\":100}"
```

```python
import requests, json, folium
```

```python
res = requests.post(url, headers = HEADER, data = DATA)
```

```python
response_json = json.loads(res.text)
```

```python
def mean(l):
    return sum(l)/len(l)
```

```python
locations = []
center = []
lat, long = [], []
for devname in response_json["rows"]:
    print(devname["latitude_deg"], devname["longitude_deg"])
    lat.append(float(devname["latitude_deg"]))
    long.append(float(devname["longitude_deg"]))
center = [mean(lat), mean(long)]
```

```python
m=folium.Map(location=[6.524400, 3.379199], tiles='cartodbpositron',zoom_start=3)
for la, lo in zip(lat, long):
    folium.Marker(location=[la, lo], popup=f"{la},{lo}",tooltip='Click here to see Popup',icon=folium.Icon(color='red',icon='off')).add_to(m)
    #folium.Marker(location=[la, lo], popup=f"{la},{lo}",tooltip='Click here to see Popup').add_to(m)
```

```python
def difference_days(t=1646780874):
    from datetime import datetime

    timestamp = t
    dt = datetime.fromtimestamp(timestamp)
    dt = str(dt)
    d = dt.split()[0].split("-")

    return d
```

```python
from datetime import date
```

```python
d = difference_days()

now = datetime.now().strftime(format="%Y-%m-%d")
now = str(now)
n = now.split("-")

dif = date(int(n[0]), int(n[1]), int(n[2])) - date(int(d[0]), int(d[1]), int(d[2]))
dif.days
```
